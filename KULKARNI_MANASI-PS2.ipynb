The stopping criterion is 'epsilon' which I have set as the maximum number of iterations to be 1000. The function here takes in arguments which are the current value of X, the learning rate and the epsilon value. I am printing the number rounded to two decimal places as I wanted the algorithm to take bigger steps. 

```python
import math
def gradient(X0, R, epsilon):
    while(epsilon>0):
        derivative=(-2*X0)
        function_value= -(X0*X0)
        X1=X0+(R*derivative)
        epsilon=epsilon-1
        print("The value of X is " , round(X1,2))
        print("The value of the gradient is", derivative)
        print("The value of the function is", function_value)
        X0=X1
        if(round(X1,2)<=0.0):
            break
        
    print("The total number of iterations are " + str(1000-epsilon))

gradient(1,0.1, 1000)
```

In short, higher learning rate results into bigger steps. And higher starting value results in lower traversing of the function.

### 1.2.7 2 D Gradient Descent


```python
# accept the quadratic eq
a = np.matrix('1,2;2,8')

# accept x y values
xy = np.matrix('2;-3')

## Define the function and gradient
## Note: you may be better of by not defining the function and using
## ufuncs instead.  But this method generalizes better.
def f(xy):
    f = -(xy.T @ a @ xy) 
    return(f)

def grad(xy):
    dv = -(a.T + a) @ xy
    return(dv)


def grad_asc_2d(xy, a, lr):
    close = 0
    iterr = 0
    x_coord = float(xy[0])
    y_coord = float(xy[1])
    X = [x_coord]
    Y = [y_coord]
    while(close<20):
        func = f(xy)
        der = grad(xy)
        xy_new = xy + lr*der
        for i in range(0, len(xy)):
            if(round(float(xy_new[i]),1) == round(float(xy[i]),1)):
               close = close + 1
            else:
               close = 0
        X.append(float(xy_new[0]))
        Y.append(float(xy_new[1]))
        xy = xy_new
        iterr = iterr + 1
        print(iterr)
    print(X)
    print(Y)
    return(X,Y)

X,Y = grad_asc_2d(xy, a, 0.1)
        
## see how it works
#x0 = np.array([1,1])
#print(f(x0))
#print(grad(x0))

## create 30x30 grid matrix
n = 30
ex1 = np.linspace(-10, 10, num=n)
ex2 = np.linspace(-10, 10, num=n)
grid1, grid2 = np.meshgrid(ex1, ex2)
## fill the grid via looping.  You may prefer to use ufuncs instead.
z = np.empty_like(grid1)
for i in range(z.shape[0]):
    for j in range(z.shape[1]):
        x = np.array([[grid1[i,j]], [grid2[i,j]]])
        z[i,j] = f(x)

       
```

From the above analysis we see that as the learning rate increases, the number of iterations decreases. This is because the algorithm takes big steps to converge to a maximum point. The speed of the convergence increases and that can be seen in the plot as well.

### Normal Distribution

```python
np.percentile(np.random.normal(size=100), 50)
```

```python
def meansamples(m, n):
    mu = np.random.normal(size=(m,n)).mean(axis=1)
    barx = mu.mean()
    std = np.sqrt(mu.var())
    x1 = barx - 1.96*std
    x2 = barx + 1.96*std
    inside = np.logical_and(mu >= x1, mu <= x2).mean()
    return (barx,std,x1, x2, inside)
meansamples(1000,10000)
```
