```python
import pandas as pd
import numpy as np
data = pd.read_csv("house-votes-84.csv", header = None)
```

```python
data = data.replace('?', 'NA')
```

```python
for elem in data.columns:
    yes = np.where(data[elem] == 'y', 1, 0)
    print("BILL NUMBER", elem, "got", np.sum(yes == 1), "yes, and", np.sum(yes == 0), "No's")
    
```

```python
# Bill 1
bill = 1
datx = data[[0,bill]]
datx_yes = datx[datx[1] == 'y']
datx_no = datx[datx[1] == 'n']
```

```python
# percentage of rep and dem in Bill 1 sayers yes
per_yes = datx_yes.groupby(0).count()/len(datx_yes)
per_yes
```

```python
# percentage of rep and dem in Bill 1 sayers no
per_no = datx_no.groupby(0).count()/len(datx_no)
per_no
```

```python
def entr(n1, n2):
    x = -n1*np.log2(n1) - n2*np.log2(n2)
    return(x)
```

```python
# entropy of bill 1 yes
x_y = entr(per_yes[1][0], per_yes[1][1])
x_y
```

```python
# entropy of bill 1 no
x_n = entr(per_no[1][0], per_no[1][1])
x_n
```

```python
#Weighted average
lens = datx.groupby(1).count()
l_no = lens[0][1]
l_yes = lens[0][2]
```

```python
# Weighted avg
w_avg = ((l_no*x_n) + (l_yes*x_y))/(l_no+l_yes)
w_avg
```

```python
# Making a function for any bill value\
def bill_entrop(billnum):
    bill = billnum
    datx = data[[0,bill]]
    datx_yes = datx[datx[bill] == 'y']
    datx_no = datx[datx[bill] == 'n']
    per_yes = datx_yes.groupby(0).count()/len(datx_yes)
    per_no = datx_no.groupby(0).count()/len(datx_no)
    x_y = entr(per_yes[bill][0], per_yes[bill][1])
    x_n = entr(per_no[bill][0], per_no[bill][1])
    lens = datx.groupby(bill).count()
    l_no = lens[0][1]
    l_yes = lens[0][2]
    w_avg = ((l_no*x_n) + (l_yes*x_y))/(l_no+l_yes)
    return(w_avg)

for i in range(1,17):
    print("The entropy of bill number", i, "is =", bill_entrop(i))
```
